```python
import numpy as np
import pickle
```

```python
np.random.randint(0,10)
```

```python
max_length = 30
```

```python
num_data = 20000
```

```python
data = []
for i in range(num_data):
    length = np.random.randint(10, max_length)
    s = [str(np.random.randint(0, 10)) for _ in range(length)]
    data.append(s)
```

```python
data = [''.join(d) for d in data]
```

```python
with open('data/algorithmic.json', 'wb') as f:
    pickle.dump(data, f)
```
